```python
#from devito import SpaceDimension, Constant, Grid, TimeFunction, Eq, Operator
#from sympy import solve
from numpy import sin, pi, linspace, shape
from examples.cfd import init_smooth

#import matplotlib.pyplot as plt

import numpy as np
from devito import Grid, Function, TimeFunction, Eq, Operator, solve
from devito import ConditionalDimension, Constant, first_derivative, second_derivative
from devito import left, right
from math import exp

import matplotlib
import matplotlib.pyplot as plt

L = 150. # Length of domain 
l = 151 # Number of points in domain
# This behaviour can be replicated for pretty much any dx
x_vals = linspace(0, L, l)


so_dev = 4 # 4th order accurate in space
to_dev = 2 # 2nd order accurate in time

extent = (L, L) # Grid is L long with l grid points
shape = (l, l)


dt = 0.2*L/(l-1)

t_end = L # Standing wave will cycle twice in time L
ns = int(t_end/dt) # Number of timesteps = total time/timestep size

grid = Grid(shape=shape, extent=extent)
time = grid.time_dim
x = grid.dimensions[0]
y = grid.dimensions[1]
h_x = grid.spacing[0]
h_y = grid.spacing[1]

# Optimized stencil coefficients
a_0 = -2.81299833
a_1 = 1.56808208
a_2 = -0.17723283
a_3 = 0.01564992

# Set up function and stencil
u_dev = TimeFunction(name="u_dev", grid=grid, space_order=so_dev, time_order=to_dev, save=ns+1)

#u_dev.data[:] = 0.1*sin(2*pi*x_vals/L)
for i in range(l):
    for j in range(l):
        u_dev.data[:,i,j] = 0.1*sin(2*pi*x_vals[i]/L)*sin(2*pi*x_vals[j]/L)


eq = Eq(u_dev.dt2-(a_3*u_dev[time, x - 3, y]
                   + a_2*u_dev[time, x - 2, y]
                   + a_1*u_dev[time, x - 1, y]
                   + a_0*u_dev[time, x, y]
                   + a_1*u_dev[time, x + 1, y]
                   + a_2*u_dev[time, x + 2, y]
                   + a_3*u_dev[time, x + 3, y])/h_x**2
        -(a_3*u_dev[time, x, y - 3]
                   + a_2*u_dev[time, x, y - 2]
                   + a_1*u_dev[time, x, y - 1]
                   + a_0*u_dev[time, x, y]
                   + a_1*u_dev[time, x, y + 1]
                   + a_2*u_dev[time, x, y + 2]
                   + a_3*u_dev[time, x, y + 3])/h_y**2)
stencil_dev = solve(eq, u_dev.forward)


bc = [Eq(u_dev[time+1,0,y], 0.0)] # Specify boundary conditions
bc += [Eq(u_dev[time+1,l-1,y], 0.0)]
bc += [Eq(u_dev[time+1,x,0], 0.0)]
bc += [Eq(u_dev[time+1,x,l-1], 0.0)]

plt.imshow(u_dev.data[0])
plt.colorbar()
plt.show()

# Create operator
op_dev = Operator([Eq(u_dev.forward, stencil_dev)] + bc)
# Apply operator
op_dev.apply(time_M=ns-1, dt=dt)


plt.imshow(u_dev.data[-1])
plt.colorbar()
plt.show()

plt.imshow(u_dev.data[-1]-u_dev.data[0])
plt.colorbar()
plt.show()
```
